2. Solve - 
a. What are the top 5 most frequently ordered products?
b. How many unique customers are there, and which customer has placed the most orders?
c. How do monthly or yearly sales trends look over time?
d. What are the peak and low sales months for each year?


```python
#a. What are the top 5 most frequently ordered products?
df['Product Name'].value_counts().head()
```

```python
#b. How many unique customers are there, and which customer has placed the most orders?
uniquecx=df['Customer Name'].nunique()
print(f'There are total {uniquecx} unique customers')
```

```python
df['Customer Name'].value_counts().head().reset_index()
```

```python
#c. How do monthly or yearly sales trends look over time?
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Year']=df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
monthlysales=df.groupby(['Year','Month'])['Sales'].sum().reset_index()
monthlysales
```

```python
#d. What are the peak and low sales months for each year?
peaksale=monthlysales.loc[monthlysales.groupby('Year')['Sales'].idxmax()]
peaksale
```

```python
lowsale=monthlysales.loc[monthlysales.groupby('Year')['Sales'].idxmin()]
lowsale
```

LINE PLOT BETWEEN YEAR AND SALES OF PEAKSALE AND LOWSALE

```python
plt.plot(peaksale['Year'],peaksale['Sales'],color='#006A67',marker='o',label='Peak sale')
plt.plot(lowsale['Year'],lowsale['Sales'],color='#AE445A',marker='*',label='Low sale')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title("Peak sale's vs. Low sale's")
plt.legend()
plt.grid()
plt.show()
```

3. Category-Based Analysis -
a. What are the top-selling categories and sub-categories?
b. How does the average order quantity vary across different categories?

```python
#a. What are the top-selling categories and sub-categories?
topcategories=df.groupby('Category')['Sales'].sum().reset_index()
topcategories
```

```python
topsubcategories=df.groupby('Sub-Category')['Sales'].sum().reset_index()
topsubcategories.head(3)
```

```python
# b. How does the average order quantity vary across different categories?
avgquantity=df.groupby('Category')['Quantity'].mean().reset_index()
avgquantity
```

4. Sales Performance by Region -
a. Which region has the highest sales, and how do sales compare across regions?
b. What is the distribution of sales across different states and cities?


```python
#a. Which region has the highest sales, and how do sales compare across regions?
topsaleregion=df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
topsaleregion.max()
```

```python
topsaleregion.reset_index()
```

```python
#b.What is the distribution of sales across different states and cities?
Statesales=df.groupby('State')['Sales'].sum().sort_values(ascending=False)
print('sales by states:',Statesales)
```

```python
plt.figure(figsize=(20,5))
plt.bar(df['State'],df['Sales'], color='#674188')
plt.xticks(rotation=90)
plt.title("sales across different states and cities")
plt.show()
```

```python
Citysales=df.groupby('City')['Sales'].sum().sort_values(ascending=False)
print('sales by citiess:',Citysales)
```
